```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y) #Functional method to plot arrays
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#Making multiple plots on 1 cell
plt.subplot(1,2,1) #column no.,row no.,plot no.
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
#OOP method to make graphs
fig = plt.figure() #Makes something like a blank canvas.

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #Adds axes

axes.plot(x,y)     #To plot

axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
#Using more of OOP
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])  
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #First two coordinates are the left bottom corner.Next two are width and height.
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("Large plot")
axes2.set_title("Smal Plot")
```

```python
#Subplots
fig,axes = plt.subplots(nrows=3,ncols=3) #To make a matrix of subplots.
plt.tight_layout() #To space the subplots in a nice manner.
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)  #Axes object is basically an array so you can interate through it and also index.
axes[0].plot(x,y)
axes[1].plot(y,x)
```

# Figure size and DPI

```python
#M1
fig = plt.figure(figsize=(8,2))  #The dimensions are width and height in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#M2
fig,axes = plt.subplots(figsize=(8,2))
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#How to save a figure
fig.savefig('enter_name_you_want.png',dpi=200)  #Enter any name and a proper extension that you want. More dpi for more detail
```

```python
#Plotting two graphs on the same figure
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
#How to use labels and legends:
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')
ax.legend()
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=5,alpha=0.5,linestyle='-',marker='o')  
#Alpha is to control transperency. You can use lw instead of linewidth. 
#Markers plot the points on the graph.
```
